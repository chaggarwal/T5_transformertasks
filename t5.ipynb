## Abstractive Summarization
Ingesting a longer piece of text and outputting a free-text summary

```python
text_to_summarize ="""
Chetan Aggarwal is a graduate of Indian Institue of Technology, Kharagpur. He completed his Integrated Master's and Bachelor's
degree in Industrial Engineering and Management in 2010. He has also pursed Actuarial Sciences from Institute of Actuaries
(India and UK) and completed 9 Actuarial exams. He started his career in the Financial Services and worked for
Fortune 500 companies like Swiss Re, AIG and GE as Data Science and Analytics professional. Currently, he is
working with Wells Fargo as Data Science Manager and leading the NLP team based out of Bengaluru, India.

Chetan grew up and finished his schooling from Chandigarh region in the northern part of India. He enjoys binge
watching movies and series, swimming, cricket. However, much of his spare time is spent in playing with his 4 year old daughter!
"""


preprocess_text = text_to_summarize.strip().replace("\n"," ")
t5_prepared_text = "summarize: "+preprocess_text
print ("original text preprocessed:\n", preprocess_text)

input_ids = tokenizer.encode(t5_prepared_text, return_tensors="pt")

input_ids
```

```python
def summarize(text):

    preprocess_text = text.strip().replace("\n","")
    t5_prepared_Text = "summarize: "+preprocess_text

    input_ids = tokenizer.encode(t5_prepared_Text, return_tensors="pt")
    
    summary_ids = model.generate(
        input_ids,
        num_beams=4,
        no_repeat_ngram_size=2,
        min_length=30,
        max_length=100,
        early_stopping=True
    )

    output = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return output
```

```python
summarize(text_to_summarize)
```

```python
def naive_sliding_window(text):

    estimated_char_per_word = 5
    windows_needed = max(1, round(len(text) / estimated_char_per_word / 512))
    print(f'Sliding over {windows_needed} windows')
    window_size = 512 * estimated_char_per_word

    for window in range(windows_needed):
        yield text[window*window_size: (window + 1)*window_size]
        
```

```python
for article_window in naive_sliding_window(article_body):  # put chunks that fit the 512 max tokens
    print(summarize(article_window))
    print()
    
```

## English -> German Translation

```python
input_ids = tokenizer.encode('translate English to German: Where is the chocolate?', return_tensors='pt')

# summmarize 
translate_ids = model.generate(
    input_ids,
    num_beams=4,
    no_repeat_ngram_size=3,
    max_length=20,
    early_stopping=True
)

output = tokenizer.decode(translate_ids[0], skip_special_tokens=True)

print (f"Translated text:\n{output}")
```

## STSB - Semantic Text Similarity Benchmark
Are two sentences semantically similar

```python
sentence_one = 'How to fish'
sentence_two = 'Fishing Manual for beginnners'


input_ids = tokenizer.encode(f"stsb sentence1: {sentence_one} sentence2: {sentence_two}", return_tensors='pt')

# summmarize 
translate_ids = model.generate(
    input_ids,
    max_length=3,
    early_stopping=True
)

output = tokenizer.decode(translate_ids[0], skip_special_tokens=True)

print(f"semantically similar? (1-5): \n{output}")

```
